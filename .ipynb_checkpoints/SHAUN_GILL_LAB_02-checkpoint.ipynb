#### Develop a model for the motion of a ball. Assuming, when on a windless day, someone standing on a bridge holds a ball over the side and tosses the ball straight up into the air. Ignore the modelling of friction (see the diagram below) in this example.

#### a)

```python
def simBallFallingNoFriction(initPos, gravity, simLength):
    t_array, v_array, s_array = [],[],[]
    v = gravity
    s = initPos
    for t in range(simLength):
        
        t_array.append(t)
        v_array.append(v)
        
        # Change in postion
        s = s + v 
        if s < 0:
            s_array.append(0)
            return t_array, v_array, s_array
        s_array.append(s)
        
        # Change in velocity
        v = v + gravity # Velocity + acceleration (gravity, no friction)

    return t_array, v_array, s_array
# No Friction
s = 10000   # Position in (m). Change in pos is r v
#v = ds/dt # Velocity, v(t) = ds/dt
#a = dv/dt # Acceleration, a(t) = dv/dt
gravity = -9.81 # (m/sec2)
simLength = 60 # seconds

t_array, v_array, s_array = simBallFallingNoFriction(s, gravity, simLength)

import pandas as pd
ballFallNoF = pd.DataFrame()
ballFallNoF.insert(loc=0, column="time(s)", value=t_array)
ballFallNoF.insert(loc=1, column="position(m)", value=s_array)
ballFallNoF
```

```python
# Kinetic friction or drag
## Estimates

### Stokes' friction, small object through air?
# F = k * v
# k, kg/sec, constant of proportionality
# v in m/sec, velocity

### Newtonian friction through air, bigger solid objects through air?
##### what defines "bigger" ?
# F = F = 0.65 * A * v^2 (v*v)
# A: Objects projected area in direction of movement PI * RADIUS^2
# v in m/sec, velocity
```

```python
import math
def simBallFallingWithFriction(m, r, v, gravity, s, simLength):
    t_array, v_array, s_array = [],[],[]
    dT = 0.001
    numIter = int(simLength / dT) + 1
    
    weight = m * gravity
    projected_area = math.pi * r * r
    
    for t in range(numIter):
        
        
        air_friction = -0.65 * projected_area * v * abs(v)
        
        total_force = weight + air_friction
        a = total_force / m 

        change_in_velocity = a

        v = v + change_in_velocity * dT # Velocity + acceleration
        
        change_in_position = v  
        
        
        
        s = s + change_in_position * dT
        
        if s < 0:
            s_array.append(0)
        else:
            s_array.append(s)
            
        t_array.append(t)
        v_array.append(v)
    return t_array, v_array, s_array

gravity = -9.81 # (m/sec2)
initV = 15
s = 400   # Position in (m)
simLength = 5 # seconds

t_array, v_array, s_array = simBallFallingWithFriction(0.5, 0.05, initV, gravity, s, simLength)

import pandas as pd
ballFallF = pd.DataFrame()
ballFallF.insert(loc=0, column="time(s)", value=t_array)
ballFallF.insert(loc=1, column="v", value=v_array)
ballFallF.insert(loc=2, column="position(m)", value=s_array)
ballFallF
```

```python
plt.plot(t_array, s_array, "g--", label="s")
plt.plot(t_array, v_array, "r--", label="v")
plt.legend()
plt.show()
```

```python
def sim(initPredPop, predBirthR, predDeathR, initPreyPop, preyBirthR, preyDeathR, dT, simLength):
    numIterations = int(simLength / dT)+1
    print(numIterations)
    tArr, predArr, preyArr = [],[],[]
    predPop = initPredPop
    preyPop = initPreyPop
    
    for i in range(1, numIterations):
        t = i * dT
        
        preyBirths = preyBirthR * preyPop
        predBirths = predBirthR * predPop  * preyPop 
        
        preyDeaths = preyDeathR * preyPop * predPop 
        predDeaths = predPop * predDeathR   
        
        
        preyPop = preyPop + (preyBirths - preyDeaths) * dT
        predPop = predPop + (predBirths - predDeaths) * dT

#       ds/dt
#         preyPop = ((preyBirthR * preyPop) - (preyDeathR * (predPop * preyPop)))
#         predArr.append(predPop)
#       dh/dt
#         predPop = (predBirthR * (preyPop * predPop)) - (predDeathR * predPop)
#         preyDeathR * preyPop
        tArr.append(t)
        predArr.append(predPop)
        preyArr.append(preyPop)
    print("done")
    return tArr, predArr, preyArr
```

```python
t, pred, prey = sim(100, 0.01, 1.06, 15, 2, 0.02, 0.001, 15)
```

```python
import matplotlib.pyplot as plt
plt.plot(t, prey, "g--", label="prey")
plt.plot(t, pred, "r--", label="pred")
plt.legend()
plt.show()
```
